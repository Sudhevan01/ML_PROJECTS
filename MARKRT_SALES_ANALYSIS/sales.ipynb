```python
ts = df.groupby('date')['prices_(£)'].mean()
print(ts.head())
```

```python
plt.figure(figsize=(12,5))
plt.plot(ts)
plt.title('Average Product Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price (£)')
plt.grid(True)
plt.show()
```

```python
rolling_mean = ts.rolling(window=7).mean()
plt.figure(figsize=(12,5))
plt.plot(ts, label='Original Prices')
plt.plot(rolling_mean, label='7-Day Rolling Mean')
plt.legend()
plt.title('Trend and Smoothing')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts, model='additive', period=7)
decomposition.plot()
plt.show()
```

```python
df = pd.read_csv('/content/All_Data_ASDA.csv', low_memory=False)
df = df.dropna(subset=['prices_(£)', 'date'])
df['date'] = pd.to_datetime(df['date'].astype(int).astype(str),format='%Y%m%d')
ts = df.groupby('date')['prices_(£)'].mean()
print(ts.head())
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(ts, order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=7)
print(forecast)
```

```python
plt.figure(figsize=(12,5))
plt.plot(ts, label='Historical Prices')
plt.plot(forecast.index, forecast, label='Forecasted Prices')
plt.xlabel('Date')
plt.ylabel('Average Price (£)')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
model = ExponentialSmoothing(ts,trend='add',seasonal=None)
fit_model = model.fit()
forecast = fit_model.forecast(7)
print(forecast)
```

```python
train = ts[:int(len(ts)*0.8)]
test = ts[int(len(ts)*0.8):]
model = ARIMA(train, order=(1,1,1))
model_fit = model.fit()
predictions = model_fit.forecast(steps=len(test))
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test, predictions)
print("MSE:", mse)
```

```python
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

```python
mean_actual = test.mean()
accuracy = (1 - (rmse / mean_actual)) * 100
print("Accuracy:", accuracy, "%")
```
